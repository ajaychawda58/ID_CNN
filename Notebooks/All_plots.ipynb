```python
def get_median(ID):
    out0, out1, out2, out3 = [], [], [], []
    out4, out5, out6, out7 = [], [], [], []
    out8, out9, out10, out11 = [], [], [], []
    for i in range(len(ID)):
        out0.append(ID[i][0])
        out1.append(ID[i][1])
        out2.append(ID[i][2])
        out3.append(ID[i][3])
        out4.append(ID[i][4])
        out5.append(ID[i][5])
        out6.append(ID[i][6])
        out7.append(ID[i][7])
        out8.append(ID[i][8])
        out9.append(ID[i][9])
        out10.append(ID[i][10])
        out11.append(ID[i][11])
    ID = [out0, out1, out2, out3, out4, out5, out6, out7, out8, out9, out10, out11]
    for j in range(len(ID)):
        ID[j] = np.median(ID[j])
    return ID
```
