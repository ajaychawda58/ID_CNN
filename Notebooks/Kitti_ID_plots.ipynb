```python
ID = np.load('../trained_model/kitti/faster_rcnn/vgg16/ID_all.npy', allow_pickle=True)
ID_channel_shift = np.load('../trained_model/kitti/faster_rcnn/vgg16/ID_all_channel_shift.npy', allow_pickle=True)
ID_horizontal_flip = np.load('../trained_model/kitti/faster_rcnn/vgg16/ID_all_horizontal_flip.npy', allow_pickle=True)
ID_horizontal_shift = np.load('../trained_model/kitti/faster_rcnn/vgg16/ID_all_horizontal_shift.npy', allow_pickle=True)
ID_rotation = np.load('../trained_model/kitti/faster_rcnn/vgg16/ID_all_rotation.npy', allow_pickle=True)
ID_vertical_flip = np.load('../trained_model/kitti/faster_rcnn/vgg16/ID_all_vertical_flip.npy', allow_pickle=True)
ID_vertical_shift = np.load('../trained_model/kitti/faster_rcnn/vgg16/ID_all_vertical_shift.npy', allow_pickle=True)
```

```python
layers = ['input', 'pool1', 'pool2', 'pool3', 'pool4', 'pool5', 'rpn_c', 'rpn_b', 'roi', 'fc', 'cls_p', 'bbox_p']
```

```python
def get_median(ID):
    out0, out1, out2, out3 = [], [], [], []
    out4, out5, out6, out7 = [], [], [], []
    out8, out9, out10, out11 = [], [], [], []
    for i in range(len(ID)):
        out0.append(ID[i][0])
        out1.append(ID[i][1])
        out2.append(ID[i][2])
        out3.append(ID[i][3])
        out4.append(ID[i][4])
        out5.append(ID[i][5])
        out6.append(ID[i][6])
        out7.append(ID[i][7])
        out8.append(ID[i][8])
        out9.append(ID[i][9])
        out10.append(ID[i][10])
        out11.append(ID[i][11])
    ID = [out0, out1, out2, out3, out4, out5, out6, out7, out8, out9, out10, out11]
    for j in range(len(ID)):
        ID[j] = np.median(ID[j])
    return ID
```

```python
def get_median_retinanet(ID):
    out0, out1, out2, out3 = [], [], [], []
    out4, out5, out6, out7 = [], [], [], []
    out8, out9 = [], []
    for i in range(len(ID)):
        out0.append(ID[i][0])
        out1.append(ID[i][1])
        out2.append(ID[i][2])
        out3.append(ID[i][3])
        out4.append(ID[i][4])
        out5.append(ID[i][5])
        out6.append(ID[i][6])
        out7.append(ID[i][7])
        out8.append(ID[i][8])
        out9.append(ID[i][9])
        
    ID = [out0, out1, out2, out3, out4, out5, out6, out7, out8, out9]
    for j in range(len(ID)):
        ID[j] = np.median(ID[j])
    return ID
```

```python
fig = plt.figure()
fig.set_size_inches(4, 4)
plt.plot(layers, ID[:,0], label='unaugmented', marker = 'o')
plt.plot(layers, ID_horizontal_shift[:,0], label='horizontal_shift', marker='+')
plt.plot(layers, ID_vertical_shift[:,0], label='vertical_shift', marker='*')
plt.plot(layers, ID_vertical_flip[:,0], label='vertical_flip', marker='s')
plt.plot(layers, ID_horizontal_flip[:,0], label='horizontal_flip',marker='x')
plt.plot(layers, ID_channel_shift[:,0], label='channel_shift',marker='d')
plt.plot(layers, ID_rotation[:,0], label='rotation',marker='p')
plt.legend()
plt.xticks(rotation=90) 
plt.savefig('4.png')
```
