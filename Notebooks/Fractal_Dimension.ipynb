```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fd_coco = pd.read_csv('../FD/coco_test.csv')
fd_coco_channel = pd.read_csv('../FD/coco_channel.csv')
fd_coco_hflip = pd.read_csv('../FD/coco_hflip.csv')
fd_coco_vflip = pd.read_csv('../FD/coco_vflip.csv')
fd_coco_rotation = pd.read_csv('../FD/coco_rotation.csv')
fd_coco_hshift = pd.read_csv('../FD/coco_hshift.csv')
fd_coco_vshift = pd.read_csv('../FD/coco_vshift.csv')
```

```python
fd_voc = pd.read_csv('../FD/voc_test.csv')
fd_voc_channel = pd.read_csv('../FD/voc_channel.csv')
fd_voc_hflip = pd.read_csv('../FD/voc_hflip.csv')
fd_voc_vflip = pd.read_csv('../FD/voc_vflip.csv')
fd_voc_rotation = pd.read_csv('../FD/voc_rotation.csv')
fd_voc_hshift = pd.read_csv('../FD/voc_hshift.csv')
fd_voc_vshift = pd.read_csv('../FD/voc_vshift.csv')
```

```python
fd_kitti = pd.read_csv('../FD/kitti_test.csv')
fd_kitti_channel = pd.read_csv('../FD/kitti_channel.csv')
fd_kitti_hflip = pd.read_csv('../FD/kitti_hflip.csv')
fd_kitti_vflip = pd.read_csv('../FD/kitti_vflip.csv')
fd_kitti_rotation = pd.read_csv('../FD/kitti_rotation.csv')
fd_kitti_hshift = pd.read_csv('../FD/kitti_hshift.csv')
fd_kitti_vshift = pd.read_csv('../FD/kitti_vshift.csv')
```

```python
def get_dim(data):
    dim = data['Dimension'][:2000]
    return dim
```

```python
coco = [fd_coco, fd_coco_hshift, fd_coco_vshift, fd_coco_hflip, fd_coco_vflip, fd_coco_channel, fd_coco_rotation]
voc = [fd_voc, fd_voc_hshift, fd_voc_vshift, fd_voc_hflip, fd_voc_vflip, fd_voc_channel, fd_voc_rotation]
kitti = [fd_kitti, fd_kitti_hshift, fd_kitti_vshift, fd_kitti_hflip, fd_kitti_vflip, fd_kitti_channel, fd_kitti_rotation]
```

```python
coco_dim, voc_dim, kitti_dim = [], [], []
for i in range(len(coco)):
    coco_dim.append(get_dim(coco[i]))
    voc_dim.append(get_dim(voc[i]))
    kitti_dim.append(get_dim(kitti[i]))    
```

```python
coco = pd.DataFrame(coco_dim)
```

```python
coco = pd.DataFrame.transpose(coco)
```

```python
coco.columns = ['normal', 'channel', 'hflip', 'vflip', 'rotation', 'hshift', 'vshift']
```

```python
voc = pd.DataFrame(voc_dim)
voc = pd.DataFrame.transpose(voc)
voc.columns = ['normal', 'channel', 'hflip', 'vflip', 'rotation', 'hshift', 'vshift']
```

```python
kitti = pd.DataFrame(kitti_dim)
kitti = pd.DataFrame.transpose(kitti)
kitti.columns = ['normal', 'channel', 'hflip', 'vflip', 'rotation', 'hshift', 'vshift']
```

```python
sns.boxplot(x="variable", y="value", data=pd.melt(coco))
plt.savefig('../Plots/FD_coco.png')
```

```python
sns.boxplot(x="variable", y="value", data=pd.melt(voc))
plt.savefig('../Plots/FD_voc.png')
```

```python
sns.boxplot(x="variable", y="value", data=pd.melt(kitti))
plt.savefig('../Plots/FD_kitti.png')
```
