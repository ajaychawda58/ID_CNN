Coco data ID calculated on model trained on VOC data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ID = np.load('ID_all_test2017.npy', allow_pickle=True)
ID_channel_shift = np.load('ID_all_channel_shift.npy', allow_pickle=True)
ID_horizontal_flip = np.load('ID_all_horizontal_flip.npy', allow_pickle=True)
ID_horizontal_shift = np.load('ID_all_horizontal_shift.npy', allow_pickle=True)
ID_rotation = np.load('ID_all_rotation.npy', allow_pickle=True)
ID_vertical_flip = np.load('ID_all_vertical_flip.npy', allow_pickle=True)
ID_vertical_shift = np.load('ID_all_vertical_shift.npy', allow_pickle=True)
```

```python
layers = ['input', 'pool1', 'pool2', 'pool3', 'pool4', 'pool5', 'rpn_c', 'rpn_b', 'roi', 'fc', 'cls_p', 'bbox_p']
```

```python
plt.plot(layers, ID[:,0], label='unaugmented')
plt.plot(layers, ID_channel_shift[:,0], label='channel_shift')
plt.plot(layers, ID_horizontal_flip[:,0], label='horizontal_flip')
plt.plot(layers, ID_horizontal_shift[:,0], label='horizontal_shift')
plt.plot(layers, ID_rotation[:,0], label='rotation')
plt.plot(layers, ID_vertical_flip[:,0], label='vertical_flip')
plt.plot(layers, ID_vertical_shift[:,0], label='vertical_shift')
plt.legend()
plt.xticks(rotation=90) 
plt.show()
```
