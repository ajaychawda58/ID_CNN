VOC data ID calculated on model trained on COCO data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ID = np.load('../trained_model/coco_voc/voc_on_coco/ID_all_JPEGImages.npy', allow_pickle=True)
ID_channel_shift = np.load('../trained_model/coco_voc/voc_on_coco/ID_all_channel_shift.npy', allow_pickle=True)
ID_horizontal_flip = np.load('../trained_model/coco_voc/voc_on_coco/ID_all_horizontal_flip.npy', allow_pickle=True)
ID_horizontal_shift = np.load('../trained_model/coco_voc/voc_on_coco/ID_all_horizontal_shift.npy', allow_pickle=True)
ID_rotation = np.load('../trained_model/coco_voc/voc_on_coco/ID_all_rotation.npy', allow_pickle=True)
ID_vertical_flip = np.load('../trained_model/coco_voc/voc_on_coco/ID_all_vertical_flip.npy', allow_pickle=True)
ID_vertical_shift = np.load('../trained_model/coco_voc/voc_on_coco/ID_all_vertical_shift.npy', allow_pickle=True)
```

```python
layers = ['input', 'pool1', 'pool2', 'pool3', 'pool4', 'pool5', 'rpn_cls_logits', 'rpn_bbox_pred', 'roi_pooling', 'fc_layers', 'pred_cls_score', 'pred_bbox_pred']
```

```python
fig = plt.figure()
fig.set_size_inches(18, 12)
plt.plot(layers, ID[:,0], label='unaugmented', marker='o')
plt.plot(layers, ID_channel_shift[:,0], label='channel_shift',marker="+")
plt.plot(layers, ID_horizontal_flip[:,0], label='horizontal_flip', marker = "*")
plt.plot(layers, ID_horizontal_shift[:,0], label='horizontal_shift', marker = "s")
plt.plot(layers, ID_rotation[:,0], label='rotation', marker = "x")
plt.plot(layers, ID_vertical_flip[:,0], label='vertical_flip', marker = "d")
plt.plot(layers, ID_vertical_shift[:,0], label='vertical_shift', marker = "p")
plt.legend()
plt.legend()
plt.xticks(rotation=90) 
fig.savefig('VOC_on_COCO.png', bbox_inches='tight')
#plt.show()
```
